## Calculating Correlation( Pearson's r)
Pearson's r:

1.First standardize each variable

2.Multiply each pair of values, and take the average

r = average of (x in std units) * (y in std units)

```python
import os
os.chdir('C:\Users\suh0902\Desktop\CP\dataset')


subway_df = pd.read_csv('nyc_subway_weather.csv')
```

```python
subway_df
```

```python
def correlation(x, y):
    std_x = (x - x.mean()) / x.std(ddof=0)
    std_y = (y - y.mean()) / y.std(ddof=0)
    return (std_x * std_y).mean()
```

```python
correlation(subway_df['ENTRIESn_hourly'], subway_df['meanprecipi'])
```

```python
correlation(subway_df['ENTRIESn_hourly'], subway_df['ENTRIESn'])
```

```python
grades_df = pd.DataFrame(
    data={'exam1': [43, 81, 78, 75, 89, 70, 91, 65, 98, 87],
          'exam2': [24, 63, 56, 56, 67, 51, 79, 46, 72, 60]},
    index=['Andre', 'Barry', 'Chris', 'Dan', 'Emilio', 
           'Fred', 'Greta', 'Humbert', 'Ivan', 'James']
)
```

```python
def convert_grade(grade):
    if grade >= 90:
        return 'A'
    elif grade >= 80:
        return 'B'
    elif grade >= 70:
        return 'C'
    elif grade >= 60:
        return 'D'
    else:
        return 'F'
```

```python
def convert_grades(grades):
    return grades.applymap(convert_grade)
```

```python
convert_grades(grades_df)
```

```python
grades_df = pd.DataFrame(
    data={'exam1': [43, 81, 78, 75, 89, 70, 91, 65, 98, 87],
          'exam2': [24, 63, 56, 56, 67, 51, 79, 46, 72, 60]},
    index=['Andre', 'Barry', 'Chris', 'Dan', 'Emilio', 
           'Fred', 'Greta', 'Humbert', 'Ivan', 'James']
)
```

```python
def convert_grades_curve(exam_grades):
    """the qcut() function here:
    http://pandas.pydata.org/pandas-docs/stable/generated/pandas.qcut.html"""
    return pd.qcut(exam_grades,[0, 0.1, 0.2, 0.5, 0.8, 1],labels=['F', 'D', 'C', 'B', 'A'])
```

```python
convert_grades_curve(grades_df['exam1'])
```
